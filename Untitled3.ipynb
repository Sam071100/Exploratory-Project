```python
import pandas as pd
df = pd.read_csv("Rigging.csv")
df.head()
```

```python
df.drop(['Sr. No.'],axis='columns', inplace=True) 
df.head()
```

```python
dummies= pd.get_dummies(df.Remarks)
dummies
```

```python
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
final = merged.drop(['Remarks'],axis='columns')
final
```

```python
inputs = final.drop('drum_rope_broke',axis='columns')
target= final.drum_rope_broke
```

```python
inputs.columns[inputs.isna().any()]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
